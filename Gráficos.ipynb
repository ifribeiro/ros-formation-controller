```python
# -*- coding: utf-8 -*-
from __future__ import unicode_literals
import math
import numpy as np
import random as rd 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_erros = pd.read_csv("Erros.csv")
```

```python
i = len(df_erros['X'])
t = np.arange(0,20,0.2)
```

```python
df_erros.head()
```

```python
fig, ax = plt.subplots(2)
fig.set_figheight(20)
fig.set_figwidth(20)
ax[0].plot(t[:i], df_erros['X'], label="Erro X")
ax[0].plot(t[:i], df_erros['Y'], label="Erro Y")        
ax[0].plot(t[:i], df_erros['Z'], label="Erro Z")
ax[0].set_title("Erros de posicionamento")

ax[1].plot(t[:i], df_erros['Rho'], '--', label="Erro rho")        
ax[1].plot(t[:i], df_erros['Beta'], '--', label="Erro beta")        
ax[1].plot(t[:i], df_erros['Alpha'], '--', label="Erro alpha")
ax[1].set_title("Erros de formação")
```
